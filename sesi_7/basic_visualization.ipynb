# Plotting

## Line Plot
menunjukkan perubahan terhadap waktu

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

# styling
mpl.style.use(['ggplot'])
```

```python
# plot garis untuk data imigran haiti
years = list(range(1980,2014))
haiti = df_can_country.loc['Haiti',years]
haiti
```

```python
haiti.plot();

plt.title('Imigration from Haiti');
plt.ylabel('number of immigrants');
plt.xlabel('Years');

plt.text(2000,6000, 'Gempa Haiti 2010');
```

**Pertanyaan 1** : Bagaimana perbandingan jumlah imigran China dan India dari tahun 1980 - 2013 ke Canada

```python
negara = ['China','India']
china_india = df_can_country.loc[negara,years]
china_india
```

```python
china_india.plot();
```

```python
# transpose
CI_transposed = china_india.T
CI_transposed.head()
```

```python
# transpose
CI_transposed.plot();
plt.title('Imigran dari China dan India ke Canada');
plt.ylabel('Jumlah Imigran');
plt.xlabel('Tahun');
```

**Pertanyaan 2** : Bandingkan tren dari 5 negara teratas yang menyumbang imigran ke Kanada

```python
top5_index = df_can_country['Total'].sort_values(ascending=False).head().index
top5_df = df_can_country.loc[top5_index, years]
top5_df = top5_df.T
```

```python
top5_df
```

```python
top5_df.plot(figsize=(18,10));
plt.title('Imigran dari 5 negara teratas');
plt.ylabel('Jumlah Imigran');
plt.xlabel('Tahun');
```

## Area Plot

```python
top5_df.plot(kind='area',
             stacked=False,
             figsize=(18,10));
plt.title('Imigran dari 5 negara teratas');
plt.ylabel('Jumlah Imigran');
plt.xlabel('Tahun');
```

## Histogram

```python
df_2013 = df_can_country[2013]
df_2013
```

```python
# numpy
count, bin_edges = np.histogram(df_2013, bins=10)

print(count)
print(bin_edges)
```

```python
df_2013.plot(kind='hist',
             bins=10,
             xticks=bin_edges,
             figsize=(8,5));
plt.title('Histogram dari data Imigran 2013')
plt.xlabel('Jumlah Imigran')
plt.ylabel('Jumlah Negara')
```

**Pertanyaan 3** : Bagaimana distribusi dari Imigran Denmark, Norwegia dan Swedia?

```python
df_dns = df_can_country.loc[['Denmark','Norway','Sweden'], years]
df_dns
```

```python
df_dns.T.plot(kind='hist',
              alpha=0.5,
              bins=15,
              figsize=(10,6),
              color=['coral','darkslateblue','mediumseagreen'])
```

## Pie Chart

```python
df_continent = df_can_country.groupby('Continent').sum()
df_continent
```

```python
df_continent['Total'].plot(kind='pie',
                           figsize=(8,9),
                           autopct='%.2f%%',
                           startangle=90,
                           shadow=True);
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continent['Total'].plot(kind='pie',
                           figsize=(15,6),
                           autopct='%.2f%%',
                           startangle=90,
                           shadow=True,
                           labels=None,
                           pctdistance=1.15,
                           explode=explode_list);

plt.legend(labels=df_continent.index);
plt.title('Data Imigrasi sesuai kontinen', y=1.12);
plt.axis('equal');
```
